```python
# Program 1: Creating DataFrame using Dictionary
import pandas as pd
data = {'apples':[3,0,1,2],'oranges':[3,0,1,2]}
df = pd.DataFrame(data)
print(df)
```

```python
# Program 2: Creating DataFrame with Custom Index
df = pd.DataFrame(data, index=["a","b","c","d"])
print(df)
```

```python
# Program 3: Creating DataFrame using from_dict()
data = {'col_1':[3,2,1,0],'col_2':['a','b','c','d']}
df = pd.DataFrame.from_dict(data)
print(df)
```

```python
# Program 4: Creating DataFrame using orient='index'
data = {'row_1':[3,2,1,0],'row_2':['a','b','c','d']}
df = pd.DataFrame.from_dict(data, orient='index')
print(df)
```

```python
# Program 5: DataFrame using orient='index' with Column Names
df = pd.DataFrame.from_dict(data, orient='index', columns=["A","B","C","D"])
print(df)
```

```python
# Program 6: Creating and Displaying DataFrame without using CSV file
import pandas as pd
data = {'Name':['A','B','C'],'Marks':[80,75,90]}
df = pd.DataFrame(data)
print(df.to_string())
```

```python
# Program 7: Removing Missing Values using dropna()
df = df.dropna()
print(df.to_string())
```

```python
# Program 8: Creating DataFrame (Cars Dataset)
data = {'cars':["BMW","Volvo","Ford"],'passings':[3,7,2]}
df = pd.DataFrame(data)
print(df)
```

```python
# Program 9: Creating Series from List
import pandas as pd
a = [1,7,2]
myvar = pd.Series(a)
print(myvar)
```

```python
# Program 10: Creating Series with Custom Index
myvar = pd.Series(a, index=["x","y","z"])
print(myvar)
print(myvar["y"])
```

```python
# Program 11: Creating Series from Dictionary
calories = {"day1":420,"day2":456,"day3":453}
myvar = pd.Series(calories)
print(myvar)
```

```python
# Program 12: Creating DataFrame (Calories and Duration)
data = {"calories":[420,380,390],"duration":[50,40,45]}
df = pd.DataFrame(data)
print(df)
```
