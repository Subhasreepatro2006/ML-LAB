```python
# PROGRAM 1: Creation of NumPy Array
import numpy as np

a = np.array([1, 2, 3, 4, 5])
print(a)
```

```python
# PROGRAM 2: Addition of Two NumPy Arrays
import numpy as np

a = np.array([5, 10, 15])
b = np.array([1, 2, 3])

print("Sum:", a + b)
```

```python
# PROGRAM 3: Multiplication of NumPy Arrays
import numpy as np

a = np.array([2, 4, 6])
b = np.array([3, 5, 7])

print("Multiplication:", a * b)
```

```python
# PROGRAM 4: Find Minimum, Maximum and Mean
import numpy as np

a = np.array([10, 20, 30, 40])

print("Minimum:", a.min())
print("Maximum:", a.max())
print("Mean:", a.mean())
```

```python
# PROGRAM 5: Creation of 2D NumPy Array
import numpy as np

a = np.array([[1, 2], [3, 4]])
print(a)
```

```python
# PROGRAM 6: Simple Line Plot using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 30]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# PROGRAM 7: Line Plot with Displayed Points
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [5, 10, 7, 14]

plt.plot(x, y, marker='o')
plt.title("Line Plot with Points")
plt.show()
```

```python
# PROGRAM 8: Line Plot Showing Start and End Points
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

plt.plot(x, y, marker='o')
plt.scatter([x[0], x[-1]], [y[0], y[-1]], color='red')
plt.title("Start and End Points")
plt.show()
```

```python
# PROGRAM 9: Bar Chart using Matplotlib
import matplotlib.pyplot as plt

x = ['A', 'B', 'C', 'D']
y = [10, 15, 7, 20]

plt.bar(x, y)
plt.title("Bar Chart")
plt.show()
```

```python
# PROGRAM 10: Bar Chart with Axis Labels
import matplotlib.pyplot as plt

x = ['Python', 'Java', 'C', 'AI']
y = [80, 70, 60, 90]

plt.bar(x, y)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Marks Bar Chart")
plt.show()
```

```python
# PROGRAM 11: Histogram Plot
import matplotlib.pyplot as plt

data = [10, 20, 10, 30, 40, 20, 50, 10, 60]

plt.hist(data, bins=5)
plt.title("Histogram")
plt.show()
```

```python
# PROGRAM 12: Pie Chart using Matplotlib
import matplotlib.pyplot as plt

sizes = [40, 20, 25, 15]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels)
plt.title("Pie Chart")
plt.show()
```

```python
# PROGRAM 13: Pie Chart with Starting Angle
import matplotlib.pyplot as plt

sizes = [30, 20, 25, 25]
labels = ['Apple', 'Banana', 'Grapes', 'Mango']

plt.pie(sizes, labels=labels, startangle=90)
plt.title("Pie Chart with Start Angle")
plt.show()
```

```python
# PROGRAM 14: Scatter Plot
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 3, 6, 2, 8]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
# PROGRAM 15: Scatter Plot Showing Start and End Points
import matplotlib.pyplot as plt

x = [2, 4, 6, 8, 10]
y = [3, 7, 5, 9, 6]

plt.scatter(x, y)
plt.scatter([x[0], x[-1]], [y[0], y[-1]], color='red')
plt.title("Scatter Plot with Start and End Points")
plt.show()
```
